```python


def binary_entropy(p):
    if p <= 0 or p >= 1:
        return 0.0
    # math.log2 works on floats; if using arrays, use np.log2
    return -(p * math.log2(p) + (1 - p) * math.log2(1 - p))

def exps_bound(Epsilon,Delta,Zeta):
    if not Delta < Epsilon*(1 - 0.75/Zeta):
        print(Epsilon - 3/(4*Zeta))
        return 0,0
    L_term = 3*Epsilon + 4*Zeta*(Delta-Epsilon)
    constant_term = 2*Epsilon*np.log(3*Delta) - 2*Delta*np.log(Delta) + 2*np.log(2) + Epsilon*np.log(Epsilon)
    return L_term,constant_term

```

```python
zeta = 0.81
epsilon = 1/4
delta = 0.01
print(epsilon - 3/(4*zeta))
term1,term2 = exps_bound(epsilon,delta,zeta)
print(term1,term2)
values = []
for L in range(1,400000):
    if term1*L + term2 <0:
        values.append(L)

print(len(values))
```
